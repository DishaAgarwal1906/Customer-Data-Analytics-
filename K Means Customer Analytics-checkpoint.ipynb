# Standardization

```python
scaler = StandardScaler()
segmentation_std = scaler.fit_transform(df_segmentation)
```

# Hierarchical Clustering

##### Mostly used when no. of data points are less as time consuming

#####  There are 2 types of clustering [Bottom Up / agglomeretive] and [Top down/Divisive] 
#####  To measure distance we have 3 methods 
##### 1.  Eulcidean Distance
##### 2.  Manhattan Distance
##### 3.  Maximum Distance

```python
hier_clust = linkage(segmentation_std, method ='ward')
```

```python
plt.figure(figsize=(10,5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hier_clust,
          show_leaf_counts=False,
          no_labels =True)
```

# K-Means Clustering 

##### 1. Choose the no. of clusters (K=2)
##### 2. Specify starting centroid
##### 3. Calculate Eucledian Distance and Rearrange the centroid
##### 4. Do this till no change can be done further

#### Issues 
##### 1. Sensitive to outliers - Solved by K-Median Clustering 
##### 2. We have to choose cluster on our own and wrong cluster leads to wrong prediction
##### 3. Elongated clusters have difficulty to get seperated

```python
## Let's find what is the optimal number of clusters which we need to add in the model 

wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state = 42)
    kmeans.fit(segmentation_std)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss, marker ='o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K Means Clustering')
plt.show()
```

```python
# By using Elbow Methos we got to know the value is 4 

kmeans = KMeans(n_clusters = 4, init='k-means++', random_state= 42)
kmeans.fit(segmentation_std)
```

```python
def_segm_kmeans = df_segmentation.copy()
def_segm_kmeans['Segment K-Means'] = kmeans.labels_
```

```python
def_segm_analysis = def_segm_kmeans.groupby(['Segment K-Means']).mean()
def_segm_analysis
```

```python
def_segm_analysis['N-Obs'] = def_segm_kmeans[['Segment K-Means','Sex']].groupby(['Segment K-Means']).count()
```

```python
def_segm_analysis['Prop Obs'] = def_segm_analysis['N-Obs']/def_segm_analysis['N-Obs'].sum()
```

```python
def_segm_analysis
```

```python
def_segm_analysis.rename({0 : 'Single Standard',
1 : 'Few Opportunity', 
2 : 'Young women Well-Off',
3 : 'Old men Well -Off'})
```

```python
def_segm_kmeans['Labels'] = def_segm_kmeans['Segment K-Means'].map({0 : 'Single Standard',
1 : 'Few Opportunity', 
2 : 'Young women Well-Off',
3 : 'Old men Well -Off'})
```

```python
x_axis = def_segm_kmeans['Age']
y_axis=def_segm_kmeans['Income']
plt.figure(figsize=(10,6))
sns.scatterplot(x=x_axis,y=y_axis,hue=def_segm_kmeans['Labels'],palette = ['g','r','c','m'])
plt.title('Segmentation K Means')
```

# Dimentionality Reduction Using PCA 
##### For better Analysis 
##### It gives same number of values as the no. of columns in decending order according to the variance 

```python
pca = PCA()
```

```python
pca.fit(segmentation_std)
```

```python
pca.explained_variance_ratio_
```

```python
# Now choose the no. of values to take 
# Generally take values upto extent that around 80% of the data could be recovered 
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,8), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle = '--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Varience')
plt.title('Explained Varience by Components')
plt.show()
```

```python
# Lets take value as 3 
pca = PCA(n_components = 3)
pca.fit(segmentation_std)
```

# PCA Results

```python
pca.components_
```

```python
df_pca_components = pd.DataFrame(data = pca.components_, 
                                columns = df_segmentation.columns.values,
                                index = ['component1','component2','component3'])
df_pca_components
```

```python
sns.heatmap(df_pca_components,
           vmin=-1,
           vmax=1,
           cmap = 'RdBu',
           annot=True)
plt.yticks([0,1,2],
          ['component1','component2','component3'],
          rotation =45,
          fontsize =9)
```

```python
scores_pca = pca.transform(segmentation_std)
```

```python
scores_pca
```

```python
wcss=[]
for i in range(1,11):
    kmeans_pca = KMeans(n_clusters=i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss, marker ='o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K Means Clustering with PCA')
plt.show()
```

```python
# take 4 number of clusters again 
kmeans_pca = KMeans(n_clusters=4, init = 'k-means++', random_state =42)
```

```python
kmeans_pca.fit(scores_pca)
```

# K Means with PCA results 

```python
df_segment_pca_kmeans = pd.concat([df_segmentation.reset_index(drop=True),pd.DataFrame(scores_pca)], axis =1)
df_segment_pca_kmeans.columns.values[-3:] = ['Component1', 'Component2','Componenet3']
df_segment_pca_kmeans['Segment KMeans PCA'] = kmeans_pca.labels_
```

```python
df_segment_pca_kmeans
```

```python
def_segm_pca_analysis = df_segment_pca_kmeans.groupby(['Segment KMeans PCA']).mean()
def_segm_pca_analysis
```

```python
def_segm_pca_analysis.rename({0 : 'Single Standard',
1 : 'Few Opportunity', 
2 : 'Young women Well-Off',
3 : 'Old men Well -Off'})
```

```python
df_segment_pca_kmeans['Labels'] = df_segment_pca_kmeans['Segment KMeans PCA'].map({0 : 'Single Standard',
1 : 'Few Opportunity', 
2 : 'Young women Well-Off',
3 : 'Old men Well -Off'})
```

```python
x_axis = df_segment_pca_kmeans['Component1']
y_axis=df_segment_pca_kmeans['Component2']
plt.figure(figsize=(10,6))
sns.scatterplot(x=x_axis,y=y_axis,hue=df_segment_pca_kmeans['Labels'],palette = ['g','r','c','m'])
plt.title('Segmentation K Means')
```
